# **简介**

## **1980 年至 2023 年的热门电子游戏数据**


简介：数据集包含从 1980 年到 2023 年的视频游戏列表，并提供发布日期、用户评论评级和评论家评论评级等信息。

变量含义：
- Title：游戏标题
- Release Date：游戏首个版本发布日期
- Team：游戏开发团队
- Rating：平均评分
- Times Listed：列出此游戏的用户数量
- Number of Reviews：用户提供的评论数量
- Genres：游戏所属的所有类型/流派
- Summary：团队提供的摘要/概述
- Reviews：用户的评价/评论

## **读取数据**

导入pandas库，使用Pandas的`read_csv`函数，将原始数据文件“games.csv”里的数据内容，解析为DatatFrame，并赋值给变量`original_data`。

```python
import pandas as pd
```

```python
original_data = pd.read_csv("../[赠送] 练习数据集（持续更新）/games.csv")
original_data
```

## **评估数据**

这一部分，我将对在上一步分建立的`original_data`这个DataFrame所包含的数据进行评估。

评估主要从两个方面进行：结构和内容，即整齐度和干净度。数据的结构性问题指不符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”
这三个标准，数据的内容性问题包括存在丢失的数据、重复数据、无效数据等等。

```python
pd.set_option("display.max_colwidth",20)
original_data.sample(10)
```

从抽样的10行数据来看，数据符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”，具体来看每行是关于某游戏的记录，每列是游戏各个属性的变量，因此不存在结构性问题

### **评估数据干净度**

```python
original_data.info()
```

从输出结果来看，数据共有1512条观察值，而`Team`、`Rating` 、`Summary`变量存在缺失值。

#### **评估缺失数据**

在了解`Team`存在缺失值后,根据条件提取出缺失观察值:

```python
original_data[original_data["Team"].isnull()]
```

`Team`表示其游戏制作公司，从输出结果来看，缺失`Team`变量的数据只有一条，并且该变量的缺失对此次数据整理并无影响，所以可以忽略。

`Rating`变量也提取观察，判断其缺失情况:

```python
original_data[original_data["Rating"].isnull()]
```

`Rating`变量是针对某款游戏的平均评分；从输出结果来看，`Rating`变量为空的数据有13条，且多数都是游戏发售日期待定或是跳票延期。
因此没有平均分和评价，所以对本次游戏统计信息没有影响，可以将`Rating`变量为空的游戏条目剔除。

`Summary`变量存在缺失值，提取并观察情况：

```python
original_data[original_data["Summary"].isnull()]
```

`Summary`变量是对该游戏的概述；从输出结果来看符合变量为空的就1条，且该游戏属于未明确发售日期，因此条该数据在本次统计中没有意义，可以将其剔除。

#### **评估重复数据**

根据数据变量来看，`Title`变量是唯一标识符，表示该游戏的名称；可能存在重复的可能性，因此将对其进行查重。

```python
original_data["Title"].duplicated()
```

```python
original_data["Team"].duplicated()
```

```python
original_data[original_data.duplicated(subset=["Team"])]
```

从输出结果可以看到，`Title`变量；游戏名称不存在重复的数据。`Team`变量；开发团队有重复的数据，但同一个公司可以开发多款游戏，因此`Team`可以存在重复。

那么针对该数据集，我们后续不需要进行去重复数据的操作。

#### **评估不一致数据**

不一致数据可能存在于`Genres`变量中，我们需要查看是否是多个不同值指代同一种游戏类型。

```python
original_data["Genres"].value_counts()
```

从输出结果来看，一款游戏可以有多种类型和风格，因此我们不会对`Genres`做后续操作。

#### **评估无效或错误数据**

```python
original_data["Rating"].describe()
```

`Rating`变量表示；游戏评分，从输出结果来看，该数据集中`Rating`变量条目中并无错误数据。

缺失的数据在上文中有过评估，并将在清理数据部分中进行处理。

## **清理数据**

根据前评估部分得到的结论，我们需要进行的数据清洗包括：


-  把`Unnamed:0`变量重名为`Data`
-  把`Data`变量设为新的索引
-  把`Rating`变量缺失的观察值删除

 为了区分开经过清理的数据和原始的数据，我们创建新的变量`cleande_data`,让它为`Original_data`复制出的副本。我们之后的清理步骤都将被运用在`cleande_data`上。

```python
cleande_data = original_data.copy()
cleande_data.head()
```

把`Unnamed:0`变量重名为`Data`：

```python
cleande_data.rename(columns={"Unnamed: 0":"Data"},inplace=True)
cleande_data
```

把`Data`变量设为新的索引：

```python
cleande_data.set_index("Data",inplace=True)
cleande_data
```

把`Rating`变量缺失的观察值删除,并查看删除后该列空缺值的和:

```python
cleande_data.dropna(subset=["Rating"],inplace=True)
```

```python
cleande_data["Rating"].isnull().sum()
```

```python
cleande_data[cleande_data["Rating"].isnull()]
```

## **保存清理后的数据**

完成数据清理后，把干净整齐的数据保存到新的文件里，文件名为`games_1980s_2023s_data_cleaned.csv`。

```python
cleande_data.head()
```
